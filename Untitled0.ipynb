```python
import os
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
from roboflow import Roboflow
```

```python
# Connect to Roboflow
rf = Roboflow(api_key="ZNmKaiIJZkAu74k0LLnC")  # Replace with your API key

# Get dataset from Roboflow
project = rf.workspace("test-2-csjzq").project("signs-n3pju")
version = project.version(2)
dataset = version.download("yolov8")
```

```python
# Define dataset path
dataset_path = dataset.location
image_folder = os.path.join(dataset_path, "train", "images")  # Adjust if needed

# Get a list of images
image_files = [f for f in os.listdir(image_folder) if f.endswith(('.jpg', '.png'))]

# Show first 5 images
num_images = min(5, len(image_files))  # Display up to 5 images
fig, axes = plt.subplots(1, num_images, figsize=(15, 5))

for i in range(num_images):
    img_path = os.path.join(image_folder, image_files[i])
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB

    axes[i].imshow(img)
    axes[i].axis("off")
    axes[i].set_title(f"Image {i+1}")

plt.show()
```

```python
# Load YOLO model
model = YOLO("yolov8n.pt")  # Use pre-trained model

# Train the model
model.train(data=os.path.join(dataset.location, "data.yaml"), epochs=50, imgsz=640)
```
